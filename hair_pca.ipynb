# 2.5) Build the covariance matrix and calculate the eigenvalues and the eigenvector

```python

cov_matrix = pd.DataFrame(np.cov(data_zscore.T))
cov_matrix
```

```python
#Extracting Eigen Values and Eigen Vectors
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
print('\n Eigen Values \n', eig_vals)
print('\n Eigen Vectors \n')
Eigen_vector = pd.DataFrame(eig_vecs)
Eigen_vector
```

# 2.6 Write the explicit form of the first PC (in terms of Eigen Vectors).

The first eigen vector represents the direction. # Discuss the cumulative values of the eigenvalues. How does it help you to decide on the optimum number of principal components? What do the eigenvectors indicate? Perform PCA and export the data of the Principal Component scores into a data frame.¶

```python
x = data_zscore.drop(["Satisfaction"],axis = 1)
y = data_zscore["Satisfaction"]
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(x,y,random_state=2,test_size = 0.3)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components= None)
```

```python
x_train_trf = pca.fit_transform(x_train)
```
